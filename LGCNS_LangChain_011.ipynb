```python
# 2) Colab 파일 업로드 모듈
from google.colab import files
import os

# 3) PDF 파일 업로드
print("📤 PDF 파일을 업로드하세요...")
uploaded = files.upload()
filename = list(uploaded.keys())[0]  # 업로드된 첫 번째 파일명

# 4) LangChain PDF 로더 불러오기
from langchain_community.document_loaders import PyPDFLoader

# 5) PDF 로딩 및 페이지 추출
loader = PyPDFLoader(filename)
pages = loader.load()

# 6) 결과 출력
print(f"📄 총 페이지 수: {len(pages)}")

n = 0  # 예: 첫 페이지 미리보기
print(f"\n📄 페이지 {n + 1} 내용:\n")
print(pages[n].page_content[:500])  # 앞 500자 미리보기
```
